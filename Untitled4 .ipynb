```python
#ARIMA (Autoregressive Integrated Moving Average)
ARIMA is a model which is used for predicting future trends on a time series data.
It is model that form of regression analysis.

1.AR (Autoregression) : Model that shows a changing variable that regresses on its own lagged/prior values.
2.I (Integrated) : Differencing of raw observations to allow for the time series to become stationary
3.MA (Moving average) : Dependency between an observation and a residual error from a moving average model
```
